```python
# Import all the required packages
import pandas as pd
import statsmodels.api as sm
import statsmodels.stats.outliers_influence as inf

# Read in the dataset and name it "cereals"
cereals = pd.read_csv(r"C:\Users\jevers\Documents\Jupyter Files\CPSC 5175\Datasets\Website Data Sets\Cereals.csv")
```

```python
X = pd.DataFrame(cereals[['Sugars', 'Fiber', 'Potass']])
pd.plotting.scatter_matrix(X)
```

```python
X = X.dropna()
```

```python
X = sm.add_constant(X)
```

```python
[inf.variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
```
