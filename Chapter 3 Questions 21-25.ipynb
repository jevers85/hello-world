```python
#Create a new column in the table called saturated_fat_per_gram
nutrition_subset['saturated_fat_per_gram'] =(nutrition_subset['saturated_fat'] / nutrition_subset['weight_in_grams']) 
# Sort the table in descending order by the saturated_fat_per_gram value
nutrition_subset_sort = nutrition_subset.sort_values(['saturated_fat_per_gram'], ascending = False)
# Display the first fifteen rows of the sorted table
nutrition_subset_sort.head(n=15)
```

<p> <em style="color:#3358FF";> 24. Standardize the field saturated_fat_per_gram. Produce a listing of all the food items that are outliers at the high end of the scale. How many food items are outliers at the low end of the scale?</em></p>

```python
# Use the zscore method from the stats class to calculate a standardized value for saturated_fat_per_gram 
nutrition_subset['saturated_fat_per_gram_z'] = stats.zscore(nutrition_subset['saturated_fat_per_gram'])

# Create a subset of the table with records whose saturated_fat_per_gram_z is greater than 3, and are 
# therefore outliers on the high side
nutrition_subset.query('saturated_fat_per_gram_z > 3' )
```

```python
# Create a subset of the data showing the outliers on the low side. Note that this doesn't return any items
nutrition_subset.query('saturated_fat_per_gram_z < -3' )
```

```python
# Because no records were shown for the low side outliers, we want to see what the lowest value actually is, so we look at the 
# tail of the sorted table. We could have also set ascending = True and accomplished the same thing by reviewing the head.
nutrition_subset_sort = nutrition_subset.sort_values(['saturated_fat_per_gram_z'], ascending = False)
nutrition_subset_sort.tail(n=5)
```

<p style="color:#2CC531"> We can see there are 15 items that are outliers on the high end of the scale. When performing the query for z-scores less than -3, no results are returned. If we review the tail of the table, we will see that the lowest value for the saturated_fat_per_gram_z z-score is -0.562335. Since this value is above -3, we would conclude there are no outliers on the low end of the data set.</p>
