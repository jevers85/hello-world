```python
# Import all the required packages
import pandas as pd
from scipy import stats
from sklearn.cluster import KMeans

# Read in the dataset and name it "cereals"
cereals = pd.read_csv(r"C:\Users\jevers\Documents\Jupyter Files\CPSC 5175\Datasets\Website Data Sets\Cereals.csv")
```

```python
# Inspect the dataset to verify it looks correct
cereals.head()
```

<strong><ol start=15><li> Subset the <i>Fat</i> and <i>Sodium</i> variables into their own data frame, X. Standardize the dataset.</ol></strong>

```python
# Isolate the predictor variables and store them
X = cereals[['Fat', 'Sodium']]

# Standardize the predictor variables
Xz = pd.DataFrame(stats.zscore(X), columns = ['Fat', 'Sodium'])
```

<strong><ol start=16><li> Run k-means clustering on the dataset, using three clusters.</strong></ol>

```python
# Run the clustering algorithm, using 3 clusters
kmeans01 = KMeans(n_clusters = 3).fit(Xz)
```

```python
# Save the cluster label information of each cluster
cluster = kmeans01.labels_
```

```python
# Separate into groups based on membership
Cluster1 = Xz.loc[cluster == 0]
Cluster2 = Xz.loc[cluster == 1]
Cluster3 = Xz.loc[cluster == 2]
```
